```python
import pandas as pd
import numpy as np
from datetime import datetime
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# LOAD DATASET

df = pd.read_csv(r"C:\Users\Admin\Desktop\PROJECT FILE\cybersecurity_intrusion_data.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

# SEVERITY CLASSIFICATION

def classify_severity(row):

    score = 0

    if 'failed_logins' in df.columns:
        score += row['failed_logins']

    if 'network_traffic' in df.columns:
        score += row['network_traffic'] / 100

    if 'login_attempts' in df.columns:
        score += row['login_attempts']

    if score < 20:
        return "Low"

    elif score < 50:
        return "Medium"

    elif score < 100:
        return "High"

    else:
        return "Critical"

df["Severity"] = df.apply(classify_severity, axis=1)

# ALERT GENERATION

alerts = []

for index, row in df.iterrows():

    severity = row["Severity"]

    alert = {
        "Time": datetime.now(),
        "Record": index,
        "Severity": severity,
        "Message": f"{severity} threat detected"
    }

    alerts.append(alert)

alerts_df = pd.DataFrame(alerts)

print("\nGenerated Alerts")
display(alerts_df.head())

# INCIDENT NOTIFICATIONS

print("\n================ INCIDENT NOTIFICATIONS ================")

for _, alert in alerts_df.iterrows():

    if alert["Severity"] in ["High", "Critical"]:

        print(
            f"[{alert['Time']}] "
            f"INCIDENT ALERT -> "
            f"{alert['Message']}"
        )
        
# CRITICAL ALERTS

critical_alerts = alerts_df.loc[
    alerts_df["Severity"] == "Critical"
]

print("Total Critical Alerts:", len(critical_alerts))

if len(critical_alerts) > 0:
    print("\nCRITICAL ALERTS FOUND!")
    display(critical_alerts.head())

# REAL-TIME ALERT MONITOR

print("\n================ REAL-TIME ALERTS ================")

for _, alert in alerts_df.head(10).iterrows():

    print(
        f"[{alert['Time']}] "
        f"{alert['Severity']} -> "
        f"{alert['Message']}"
    )

# EMAIL ALERT FUNCTION

def send_email_alert(subject, body,
                     sender_email,
                     sender_password,
                     receiver_email):

    try:

        message = MIMEMultipart()

        message["From"] = sender_email
        message["To"] = receiver_email
        message["Subject"] = subject

        message.attach(MIMEText(body, "plain"))

        server = smtplib.SMTP("smtp.gmail.com", 587)

        server.starttls()

        server.login(
            sender_email,
            sender_password
        )

        server.sendmail(
            sender_email,
            receiver_email,
            message.as_string()
        )

        server.quit()

        print("Email Alert Sent Successfully!")

    except Exception as e:
        print("Email Error:", e)

# SEND EMAIL FOR CRITICAL ALERTS

if len(critical_alerts) > 0:

    body = f"""
Cybersecurity Alert

Critical threats detected.

Number of Critical Alerts:
{len(critical_alerts)}

Immediate investigation required.
"""

# ALERT SUMMARY

summary = alerts_df["Severity"].value_counts()

print("\n================ ALERT SUMMARY ================")
print(summary)

# Save alerts

alerts_df.to_csv(
    "alert_management_report.csv",
    index=False
)

print("\nReport Saved:")
print("alert_management_report.csv")
```
